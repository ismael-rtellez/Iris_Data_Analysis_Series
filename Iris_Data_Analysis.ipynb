## Problem 1: Data Adquisition

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

# Create DataFrame x with feature values
x = pd.DataFrame(
    data=iris.data,
    columns=["sepal_length", "sepal_width", "petal_length", "petal_width"]
)

# Create DataFrame y with target values (species)
y = pd.DataFrame(
    data=iris.target,
    columns=["Species"]
)
```

## Problem 2: Combining Data

```python
df = pd.concat([x, y], axis=1)
df
```

## Problem 4: Examining the iris dataset itself

*   The Iris dataset is one of the oldest and most well-know datasets in pattern recognition and statistics. It was introduced by the British statistician and biologist Ronald Fisher in 1936 in his paper "The use of multiple measurements in taxonomic problems". The data was collected by Edgar Anderson, who measured the physical characteristics of iris flower from three species.
*   The dataset contains 150 observations of iris flower, with 50 samples from each of three species: `Iris setosa`, `Iris versicolor`, and `Iris virginica`. It's frequently used for classification tasks and is often considered the "hello world" dataset for machine learning.

### Difference detween .loc and .iloc:

*   `loc`: Label-based indexing. Uses the actual row/column labels to select data.
*   `iloc`: Integer position-based indexing. Use numerical indices (0, 1, 2...) to selct data.

**Example:**
*   `df.loc[10, 'sepal_width']` selects the value at row label 10, column 'sepal_width'
*   `df.iloc[10, 1]` selects value at the 11th row, 2nd column



## Problem 6: Creating a diagram

### Pie chart of species distribution

```python
import matplotlib.pyplot as plt

# Convert numerical species to actual names
species_names = {0: 'Iris setosa', 1: 'Iris versicolor', 2: 'Iris virginica'}
df['species_name']= df['Species'].map(species_names)

# Pie chart of sample counts per label
plt.figure(figsize=(10, 6))
species_counts = df['Species'].value_counts()
plt.pie(species_counts, labels=[species_names[i] for i in species_counts.index], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Distribution of Irris Species')
```

### Boxplot for each feature by species

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features, 1):
  plt.subplot(2, 2, i)
  sns.boxplot(x='Species', y=feature, data=df)
  plt.title(f'{feature.title()} Distribution')

plt.tight_layout()
plt.show()
```

### Violin Plots feature distribution by Species

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features, 1):
  plt.subplot(2, 2, i)
  sns.violinplot(x='Species', y=feature, data=df, inner= 'quartile')
  plt.title(f'{feature.title()} Distribution')

plt.tight_layout()
plt.show()
```

### **Differences between Box Plot and Violin Plot:**
*   **Box Plot:** Display the median, quartiles, and potential outliers. it provides a summary of the data distribution.
*   **Violin Plot:** Combines box plot features with a density plot, showing shape of the data. It provides more information about the data distribution.



## Problem 7: Confirming the relationship between features

### Create scatter plots of all combinations (6) such that 2 out 4 features are selected by color:

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

fig.suptitle('- Create scatter plots of all combinations (6) such that 2 out 4 features are selected by color:')

sns.scatterplot(ax=ax[0, 0], data=df, x='sepal_length', y='sepal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0, 1], data=df, x='sepal_length', y='petal_length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0, 2], data=df, x='sepal_length', y='petal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1, 0], data=df, x='sepal_width', y='petal_length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1, 1], data=df, x='sepal_width', y='petal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1, 2], data=df, x='petal_length', y='petal_width', hue='Species', palette="deep")

sns.set()
```

### Create a scatterplot matrix that displays all feature combination at once:

```python
# Scatter plot matrix
sns.pairplot(df, hue='Species', height=2.5)
```

### Correlation coefficient matrix of 4 features:

```python
# Correlation matrix
correlation_matrix = df.corr()
correlation_matrix
```

### heatmap of the correlation matrix

```python
# Heatmapp of the correlation matrix

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Correlation Coefficient Matrix')
plt.show()
```

After visualizing the data, we can draw several conclusions:
*  The scatter plot matrix reveals distinc cluster for the threeiris species, indicating that they can be separated base on their features.
*  The box plots show that `Iris setosa` has a smaller range for both sepal and petal dimensions compared to `Iris virginica` and `Iris versicolor`, which have overlapping ranges.
*  The violin plot provide a clearer view of the distribution of each feature across species, highlighting the density of data points and showing that `Iris setosa` has a more concentrated distribution.
